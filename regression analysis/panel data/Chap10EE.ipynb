```python
PooledOLSclustered = smf.ols("dem_ind ~ log_gdppc", data=df)
results = PooledOLSclustered.fit(cov_type='cluster', cov_kwds={'groups': df[['country', 'year', 'dem_ind', 'log_gdppc']].dropna()['country']})
results.summary()
```
